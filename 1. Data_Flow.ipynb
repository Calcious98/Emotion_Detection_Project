```python
import os
def convert_size(size_bytes):
    import math
    if size_bytes == 0:
        return "0B"
    size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    s = round(size_bytes / p, 2)
    return "%s %s" % (s, size_name[i])
```

```python
from PIL import Image
from autocrop import Cropper

cropper = Cropper()

if not os.path.exists('cropped_image'):
    os.makedirs('cropped_image')
if not os.path.exists('new_image'):
    os.makedirs('new_image')

for pic in file_list:
  # Get a Numpy array of the cropped image
  # print(pic)
  cropped_array = cropper.crop('neutral/'+pic)
  # print(type(cropped_array))
  if str(type(cropped_array)) ==  "<class 'numpy.ndarray'>":
    # Save the cropped image with PIL if a face was detected:
    if cropped_array.any():
      image = Image.open("/content/neutral/{}".format(pic))
      image.save("new_image/{}".format(pic))
      cropped_image = Image.fromarray(cropped_array)
      cropped_image.save('cropped_image/'+'cropped_{}'.format(pic))
  elif str(type(cropped_array)) ==  "<class 'NoneType'>":
    print(pic)
    continue
```

```python
# train, validation, test 이미지가 들어있는 폴더 경로 지정
train_dir = 'split/train' 
valid_dir = 'split/val'
test_dir  = 'split/test'
```

### 4) ImageDataGenerator( ) & flow_from_directory( )

#### Normalization
- ImageDataGenerator()
  - 이미지를 normalization 해준다 

#### Resizing & Generator
- flow_from_directory()
  - 이미지 크기를 학습에 용이한 크기로 바꿔준다 
  - batch_size를 정해줌으로 학습의 속도를 조절해 준다
  - class_mode를 다중분류로 세팅해 준다  

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

batch_size = 64
img_size = 150

# 이미지의 크기를 같게 만들어 준다 With normalization
# 각 pixel에 1/255를 곱함으로써 0~1 사이로 범위를 조정해 준다 
train_datagen = ImageDataGenerator(rescale = 1./255
                                   )
valid_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
                  train_dir,
                  target_size = (img_size,img_size),
                  batch_size = batch_size, 
                  class_mode = 'categorical')

valid_generator = valid_datagen.flow_from_directory(
                  valid_dir,
                  target_size = (img_size,img_size),
                  batch_size = batch_size,
                  class_mode = 'categorical')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

batch_size = 32
img_size = 250

# 이미지의 크기를 같게 만들어 준다 With normalization
# 각 pixel에 1/255를 곱함으로써 0~1 사이로 범위를 조정해 준다 
train_datagen = ImageDataGenerator(rescale = 1./255
                                   )
valid_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
                  train_dir,
                  target_size = (img_size,img_size),
                  batch_size = batch_size, 
                  class_mode = 'categorical')

valid_generator = valid_datagen.flow_from_directory(
                  valid_dir,
                  target_size = (img_size,img_size),
                  batch_size = batch_size,
                  class_mode = 'categorical')
```
